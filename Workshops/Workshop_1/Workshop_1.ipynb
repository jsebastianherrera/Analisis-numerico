Sqrt algorithm using Babilonia methodology

```python
n = 7
e = 1
x = 7

def calculate(x, n):
    try:
        return (1/2)*(x+(n/x))
    except:
        print("Error: Float division by zero")
        exit(0)


y = calculate(x, n)
while abs(x-y) > float(e):
    print("Y:", y)
    x = y
    y = calculate(x, n)
```
